```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from prophet import Prophet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from time import time
```

```python
df = pd.read_csv('datasets/AEP_hourly.csv')
df['Datetime'] = pd.DatetimeIndex(df.Datetime)
df.set_index('Datetime',inplace=True)
df['Date'] = df.index.date
df = df.groupby('Date')[['AEP_MW']].sum()
df.index = pd.DatetimeIndex(df.index)
df.drop(index=df.index[-1],inplace=True)
df.rename(columns={'AEP_MW':'y'},inplace=True)
df['ds'] = df.index
df.head()
```

```python
df_train = df[(df.index < pd.Timestamp(day=1,month=1,year=2018)) & (df.index > pd.Timestamp(day=1,month=1,year=2013)) ]
df_test  = df[df.index >= pd.Timestamp(day=1,month=1,year=2018)]

fig,ax = plt.subplots(figsize=(15,3),dpi=100)
plt.plot(df_train[['y']],color='k',lw=1,label='Training data')
plt.plot(df_test[['y']],color='r',lw=1,label='Test data')
plt.legend()
plt.grid(alpha=0.3)
```

```python
t1 = time()
m = Prophet()
m.fit(df_train)
t2 = time()
fitting_time = t2 - t1 
```

```python
t1 = time()
forecast = m.predict(df_test[['ds']])
t2 = time()
prediction_time = t2-t1
forecast[['yhat']].tail()
```

```python
fig,ax = plt.subplots(figsize=(15,3),dpi=100)
plt.plot(forecast[['ds']],forecast[['yhat']],color='k',lw=1)
plt.plot(df_test[['ds']],df_test[['y']],color='r',lw=1)
plt.grid(alpha=0.2)
```

```python
fig2 = m.plot_components(forecast)
```

```python
error = mean_squared_error(forecast[['yhat']],df_test[['y']],squared=False)
```

```python
print(f'Fit time        = {fitting_time:.4f}s')
print(f'Prediction time = {prediction_time:.4f}s')
print(f'MSE             = {error:,.0f}')
```
